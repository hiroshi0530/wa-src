## 三角関数

### np.sin(x)
$\sin x$です。

```python
print(np.sin(0))
print(np.sin(np.pi / 2))
print(np.sin(np.pi))
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.sin(x)

plt.grid()
plt.title('$y = \sin x$', fontsize=16)
plt.ylabel('$\sin x$')
plt.plot(x,y)
```

### np.cos(x)
$\cos x$です。

```python
print(np.cos(0))
print(np.cos(np.pi / 2))
print(np.cos(np.pi))
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.cos(x)

plt.grid()
plt.title('$y = \cos x$', fontsize=16)
plt.ylabel('$\cos x$')
plt.plot(x,y)
```

### np.tan(x)
$\tan x$です。

```python
print(np.tan(0))
print(np.tan(np.pi / 4))
print(np.tan(np.pi))
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.tan(x)

plt.grid()
plt.title('$y = \\tan x$', fontsize=16)
plt.ylabel('$\tan x$')
plt.ylim(-5,5)
plt.plot(x,y)
```

### np.arcsin(x)
$\sin x$の逆関数です。

```python
print(np.arcsin(0))
print(np.arcsin(1))
print(np.arcsin(-1))
```

```python
x = np.linspace(-1, 1, 100)
y = np.arcsin(x)

plt.grid()
plt.title('$y = \\arcsin x$', fontsize=16)
plt.plot(x,y)
```

### np.arccos(x)

```python
print(np.arccos(0))
print(np.arccos(1))
print(np.arccos(-1))
```

```python
x = np.linspace(-1, 1, 100)
y = np.arccos(x)

plt.grid()
plt.title('$y = \\arccos x$', fontsize=16)
plt.plot(x,y)
```

### np.arctan(x)
$\tan x$の逆関数です。

```python
print(np.arctan(0))
print(np.arctan(1))
print(np.arctan(-1))
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.arctan(x)

plt.grid()
plt.title('$y = \\arctan x$', fontsize=16)
plt.plot(x,y)
```

### np.sinh(x)
双曲線正弦関数です。
<div>
$
\displaystyle \sinh x = \frac{e^x - e^{-x}}{2}
$
</div>

```python
print(np.sinh(0))
print(np.sinh(-1))
print(np.sinh(1))
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.sinh(x)

plt.grid()
plt.title('$y = \sinh x$', fontsize=16)
plt.plot(x,y)
```

### np.cosh(x)
双曲線余弦関数です。
<div>
$
\displaystyle \cosh x = \frac{e^x + e^{-x}}{2}
$
</div>

```python
print(np.cosh(0))
print(np.cosh(-1))
print(np.cosh(1))
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.cosh(x)

plt.grid()
plt.title('$y = \cosh x$', fontsize=16)
plt.plot(x,y)
```

### np.tanh(x)
双曲線正接関数です。
<div>
$
\displaystyle \tanh x = \frac{\sinh x}{\cosh x}
$
</div>

深層学習の活性化関数に利用される事があります。

```python
print(np.tanh(0))
print(np.tanh(-1))
print(np.tanh(1))
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.tanh(x)

plt.grid()
plt.title('$y = \\tanh x$', fontsize=16)
plt.plot(x,y)
```

### np.arcsinh(x)
$\sinh x$の逆関数です。

```python
print(np.arcsinh(0))
print(np.arcsinh(1))
print(np.arcsinh(-1))
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.arcsinh(x)

plt.grid()
plt.title('$y = \\arcsinh x$', fontsize=16)
plt.plot(x,y)
```

### np.arccosh(x)
$\cosh x$の逆関数です。

```python
print(np.arccosh(1))
```

```python
x = np.linspace(1, np.pi, 100)
y = np.arccosh(x)

plt.grid()
plt.title('$y = \\arccosh x$', fontsize=16)
plt.plot(x,y)
```

### np.arctanh(x)
$\tanh x$の逆関数です。

```python
print(np.arctanh(0))
print(np.arctanh(0.5))
print(np.arctanh(-0.5))
```

```python
x = np.linspace(-0.99, 0.99, 100)
y = np.arctanh(x)

plt.grid()
plt.title('$y = \\arctanh x$', fontsize=16)
plt.plot(x,y)
```

### np.deg2rad(x)
弧度法からラジアン表記に変換します。

```python
np.deg2rad(45) # => pi / 4 
```

### np.rad2deg(x)
弧度法からラジアン表記に変換します。

```python
np.rad2deg(np.pi / 4)
```
